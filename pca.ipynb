# Import libraries and set up example

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA as pca

circle_1 = np.random.multivariate_normal([1.0, 1.0], [[0.5, 0.0],[0.0, 0.5]], 500)
circle_2 = np.random.multivariate_normal([-1.0, -1.0], [[0.5, 0.0],[0.0, 0.5]], 500)
dist = np.concatenate((circle_1, circle_2), axis=0)
```

# Use package **sklearn**

```python
pca_fit = pca(n_components=2).fit(dist)

##Just to plot##
x = np.linspace(-3,3,50)
pca1 = x*pca_fit.components_[0][0]/pca_fit.components_[0][1]
pca2 = x*pca_fit.components_[1][0]/pca_fit.components_[1][1]

plt.plot(x, pca1, color='red', linestyle='--')
plt.plot(x, pca2, color='red', linestyle='--')
plt.scatter(circle_1[:,0], circle_1[:,1])
plt.scatter(circle_2[:,0], circle_2[:,1])
```

# Use package **NumPy**

```python
eigen_value, eigen_vector = np.linalg.eig(np.cov(m=dist.T));

##Just to plot##
x = np.linspace(-3,3,50)
pca1 = x*eigen_vector[0][0]/eigen_vector[0][1]
pca2 = x*eigen_vector[1][0]/eigen_vector[1][1]

plt.plot(x, pca1, color='red', linestyle='--')
plt.plot(x, pca2, color='red', linestyle='--')
plt.scatter(circle_1[:,0], circle_1[:,1])
plt.scatter(circle_2[:,0], circle_2[:,1])
```

# Plot the projection onto first components of pca

```python
proj1 = np.inner(pca_fit.components_[0],circle_1)
proj2 = np.inner(pca_fit.components_[0],circle_2)
plt.scatter(pca_fit.components_[0][0] * proj1, pca_fit.components_[0][1] * proj1, color='tab:blue')
plt.scatter(pca_fit.components_[0][0] * proj2, pca_fit.components_[0][1] * proj2, color='tab:orange')
```
